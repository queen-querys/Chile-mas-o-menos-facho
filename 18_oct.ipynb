```python
import pandas as pd

#Leer el archivo CSV de encuesta 2018
encuesta_2018 = pd.read_csv('latino_2018.csv')

# Mostrar las primeras filas del dataset
print(encuesta_2018.head())
```

```python
#Leer el archivo CSV de encuesta 2023
encuesta_2023 = pd.read_csv('latino_2023.csv')

# Mostrar las primeras filas del dataset
print(encuesta_2023.head())
```

```python
#Filtrar los valores no válidos en ambos datasets
encuesta_2018_filtrada = encuesta_2018[(encuesta_2018['P22ST'] >= 0) & (encuesta_2018['P22ST'] <= 10)]
encuesta_2023_filtrada = encuesta_2023[(encuesta_2023['P16ST'] >= 0) & (encuesta_2023['P16ST'] <= 10)]
```

```python
# Contar las respuestas para cada año
conteo_2018 = encuesta_2018_filtrada['P22ST'].value_counts().sort_index()
conteo_2023 = encuesta_2023_filtrada['P16ST'].value_counts().sort_index()
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Crear los ejes para el gráfico
labels = conteo_2018.index  # Las categorías de la escala de 0 a 10
x = np.arange(len(labels))  # Posiciones en el eje X

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))

# Barras para 2018
plt.bar(x - 0.2, conteo_2018, width=0.4, label='2018', color='blue')

# Barras para 2023
plt.bar(x + 0.2, conteo_2023, width=0.4, label='2023', color='green')

# Añadir etiquetas y título
plt.xticks(x, labels)
plt.xlabel('Escala Ideológica (0 = Izquierda, 10 = Derecha)')
plt.ylabel('Frecuencia')
plt.title('Comparación de Escala Izquierda-Derecha entre 2018 y 2023')
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
